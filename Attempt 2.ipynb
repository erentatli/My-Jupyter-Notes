```python
import requests
from bs4 import BeautifulSoup

web_page = "http://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168"
page = requests.get(web_page)
soup = BeautifulSoup(page.content,'html.parser')

seven_day = soup.find(id="seven-day-forecast-container")
forecast_items = seven_day.find_all(class_="forecast-tombstone")

#today = forecast_items[0]
#period = today.find(class_="period-name").get_text()
#short_desc = today.find(class_="short-desc").get_text()
#temp = today.find(class_="temp").get_text()
#img = today.find('img')

period_names = []
short_descs = []
temps = []
descs =[]

for item in forecast_items:
    period_names.append(item.find(class_="period-name").get_text())
    short_descs.append(item.find(class_="short-desc").get_text())
    temps.append(item.find(class_="temp").get_text())
    descs.append(item.img["title"])
period_names,short_descs,temps,descs
    
```

```python
import pandas as pd
weather = pd.DataFrame({"period":period_names,"short description":short_descs,"temperature":temps,"description":descs})
weather
```
